```python
import matplotlib.pyplot as plt
import pandas as pd

"""
df = pd.read_csv("data/GOOG.csv")

y = df['Close'].head(20)
x = df['Date'].head(20)

plt.xlabel("Date")
plt.xticks(rotation = 65)
plt.ylabel("Close ($)", fontsize=18)
plt.plot(x, y)

plt.show()
"""

df = pd.read_csv("data/GOOG.csv")
print(df)

# what is the rolling mean (moving average) of the stock for the last 60 days?

# drop existing columns from dataframe
df = df.drop(columns=['Open', 'High', 'Low', 'Volume']).copy()
# limit calculation to first 50 rows
df = df.head(50).copy()

close_px = df['Close']
# use groups of 5 to calculate rolling average
mavg = close_px.rolling(window=5).mean()

print(df)
print(f"Rolling Average\n {mavg}")


close_px.plot(label='GOOG')
mavg.plot(label='avg')
plt.legend()
plt.show()
```
